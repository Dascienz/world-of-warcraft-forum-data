```python
data = pd.read_pickle(os.getcwd() + '/pickle/wow_player_stats.pkl')
data.columns = ['player','lvl_race_class','achievements','min_date','max_date']
```

### Feature Cleaning

```python
#faction keys
factions = {'.*Human.*': 'Alliance',
            '.*Dwarf.*': 'Alliance',
            '.*Gnome.*': 'Alliance',
            '.*Night Elf.*': 'Alliance',
            '.*Draenei.*': 'Alliance',
            '.*Worgen.*': 'Alliance',
            '.*Orc.*': 'Horde',
            '.*Troll.*': 'Horde',
            '.*Undead.*': 'Horde',
            '.*Tauren.*': 'Horde',
            '.*Goblin.*': 'Horde',
            '.*Blood Elf.*': 'Horde',
            '.*Pandaren.*': 'Pandaren'}

#race keys
races = {'.*Human.*': 'Human',
         '.*Dwarf.*': 'Dwarf',
         '.*Gnome.*': 'Gnome',
         '.*Night Elf.*': 'Night Elf',
         '.*Draenei.*': 'Draenei',
         '.*Worgen.*': 'Worgen',
         '.*Orc.*': 'Orc',
         '.*Troll.*': 'Troll',
         '.*Undead.*': 'Undead',
         '.*Tauren.*': 'Tauren',
         '.*Goblin.*': 'Goblin',
         '.*Blood Elf.*': 'Blood Elf',
         '.*Pandaren.*': 'Pandaren'}

#class keys
classes = {'.*Warrior.*':'Warrior',
           '.*Paladin.*':'Paladin',
           '.*Hunter.*':'Hunter',
           '.*Rogue.*':'Rogue',
           '.*Priest.*':'Priest',
           '.*Death Knight.*':'Death Knight',
           '.*Shaman.*':'Shaman',
           '.*Mage.*':'Mage',
           '.*Warlock.*':'Warlock',
           '.*Monk.*':'Monk',
           '.*Druid.*':'Druid',
           '.*Demon Hunter.*':'Demon Hunter'}
```

```python
#level column
data['level'] = data['lvl_race_class'].str.split(' ', expand=True)[0]

#race and class column
data['race-class'] = data['lvl_race_class'].str.findall('[A-Z][a-zA-Z ]+').apply(lambda x: x[0])

#race
data['race'] = data['race-class'].replace(races,regex=True)

#class
data['class'] = data['race-class'].replace(classes,regex=True)

#faction column
data['faction'] = data['race'].replace(factions,regex=True)
```

```python
data.to_pickle(os.getcwd() + '/pickle/wow_player_stats_cleaned.pkl')
```
